```python
import matplotlib.pyplot as plt
import pile1, pile2, pile3, time
import file1, file2, file3
```

```python
def trace_empiler(n):
    """Fonction permettant de comparer le temps d'exécution de l'implémentation d'une pile par une liste"""
    temps_pile1=[]
    temps_pile2=[]
    temps_pile3=[]
    liste_n = [i for i in range(1, n+1, 1)]
    for i in range(1, n+1, 1):
        ma_pile=pile1.pile()
        debut = 1000000*time.time()
        t = pile1.empiler(ma_pile,i)
        fin = 1000000*time.time()
        temps_pile1.append(fin - debut)
    for i in range(1, n+1, 1):
        ma_pile=pile2.Pile()
        debut = 1000000*time.time()
        t = ma_pile.empiler(i)
        fin = 1000000*time.time()
        temps_pile2.append(fin - debut)
    for i in range(1, n+1, 1):
        ma_pile=pile3.pile()
        debut = 1000000*time.time()
        t = pile1.empiler(ma_pile,i)
        fin = 1000000*time.time()
        temps_pile3.append(fin - debut)
    plt.plot(liste_n, temps_pile1, 'bo', label='implémentation utilisant une liste')
    plt.plot(liste_n, temps_pile2, 'ro', label='implémentation utilisant une classe')
    plt.plot(liste_n, temps_pile3, 'ko', label='implémentation utilisant deque')
    plt.legend(loc='upper left')
    plt.show()
```

```python
trace_empiler(10000)
```

```python
def trace_depiler(n):
    """Fonction permettant de comparer le temps d'exécution de l'implémentation d'une pile par une liste"""
    temps_pile1=[]
    temps_pile2=[]
    temps_pile3=[]
    liste_n = [i for i in range(1, n+1, 1)]
    for i in range(1, n+1, 1):
        ma_pile=pile1.pile()
        for i in range(0, n+1):
            pile1.empiler(ma_pile, 1)
        debut = 1000000*time.time()
        pile1.depiler(ma_pile)
        fin = 1000000*time.time()
        temps_pile1.append(fin - debut)
        
    for i in range(1, n+1, 1):
        ma_pile=pile2.Pile()
        for i in range(0, n+1):
            ma_pile.empiler(1)
        debut = 1000000*time.time()
        ma_pile.depiler()
        fin = 1000000*time.time()
        temps_pile2.append(fin - debut)
        
    for i in range(1, n+1, 1):
        ma_pile=pile3.pile()
        for i in range(0, n+1):
            pile3.empiler(ma_pile, 1)
        debut = 1000000*time.time()
        pile3.depiler(ma_pile)
        fin = 1000000*time.time()
        temps_pile3.append(fin - debut)
        
    plt.plot(liste_n, temps_pile1, 'bo', label='implémentation utilisant une liste')
    plt.plot(liste_n, temps_pile2, 'ro', label='implémentation utilisant une classe')
    plt.plot(liste_n, temps_pile3, 'ko', label='implémentation utilisant deque')
    plt.legend(loc='upper left')
    plt.show()
```

```python
trace_depiler(5000)
```
